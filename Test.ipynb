```python
from pixelink import PixeLINK
import numpy as np
import time
```

```python
cam=PixeLINK()
cam.streaming=True
cam.roi=[0,0,1280,1024]
cam.frame_rate=10
cam.shutter = 0.01

binning_conf = [[2,0,2,2],[2,1,2,2],[2,2,2,2]]
```

```python
avg_intensity = np.zeros((3,3))

for i, b_conf in enumerate(binning_conf):
    cam.streaming = False
    cam.binning = b_conf
    cam.streaming = True
    image = cam.grab()
    avg_intensity[i][0] = np.mean(image) 
    avg_intensity[i][1] = image.shape[0]
    avg_intensity[i][2] = image.shape[1]
    
print(avg_intensity)
```

```python
frame_rate_matrix = np.zeros((3,60))
exp_frame_rate_matrix = np.zeros((3,60))
frame_rate_values = np.linspace(30, 0.5, 60)

for i, b_conf in enumerate(binning_conf):
    cam.binning = b_conf
    for j, f_value in enumerate(frame_rate_values):
        cam.streaming = False
        try:
            cam.frame_rate = f_value
        except:
            print('error')
        finally:
            frame_rate_matrix[i][j] = cam.frame_rate
        print(frame_rate_matrix[i][j])
        cam.streaming = True
        cam.grab()
        t = time.time()
        cam.grab()
        exp_frame_rate_matrix[i][j] = 1/(time.time()-t)
        print(exp_frame_rate_matrix[i][j])
```

```python
print(frame_rate_matrix)
print('\n-----------------------------------')
print(exp_frame_rate_matrix)
```
